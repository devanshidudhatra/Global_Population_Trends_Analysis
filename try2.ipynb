```python
# importing all library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset1.csv")
```

```python
print(df.head)
print(df.shape)
print(df.info)
```

```python
# we need to remove some columns
df.drop(columns=['Country Code' , 'Indicator Name' ] , inplace = True)

# we see the total population of each country
total_population=df[df['Indicator Code']=='SP.POP.TOTL']
print(total_population)
```

```python
# here we are sorting data based on 2022
# one with highest total population in 2022 is printed first
sort_by_2022 = total_population.sort_values(by='2022',ascending=False)
print(sort_by_2022)
```

```python
# get top 10 countries with the highest population for 2022
top_ten_countries = sort_by_2022.head(10)
print("Top Ten Countries Of Total Population:-")
print(top_ten_countries)
```

```python
def add_value_labels(ax, bars, spacing=10):
    for bar in bars:
        x = bar.get_width()
        y = bar.get_y() + bar.get_height() / 2

        va = 'center'
        label = '{:.0f}'.format(x)  # Format the label to display as an integer
        x_shift = spacing * (1 if x >= 0 else -1)
        ax.annotate(label, (x, y), xytext=(x_shift, 0), textcoords="offset points", ha='center', va=va)

```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.figure(figsize=(20,13))
plt.rcParams['axes.facecolor'] = "black"

plt.subplot(2,2,1)
bars1 = plt.barh(y=top_ten_countries['Country Name'], width=top_ten_countries['2022'], height=0.8, color="skyblue")
plt.title("Top Ten countries Of Total Population (2022)")
plt.xlabel("Total population")
plt.ylabel("Country")
add_value_labels(plt.gca(), bars1)

plt.subplot(2,2,2)
bars2 = plt.barh(y=top_ten_countries['Country Name'], width=top_ten_countries['2000'], height=0.8, color="lightgreen")
plt.title("Top Ten countries Of Total Population in 2000")
plt.xlabel("Total population")
plt.ylabel("Country")
add_value_labels(plt.gca(), bars2)

plt.tight_layout()
plt.show()
```
